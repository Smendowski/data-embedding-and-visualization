### Reuters

```python
df_reuters = pd.read_csv('datasets_with_labels/reuters.csv', header=None)
document_identifiers = df_reuters.iloc[:, -1].unique()
print(f"Original Reuters dataset has samples for {len(document_identifiers)} unique documents. \n")
document_identifiers
```

```python
ids_to_remove = document_identifiers[
    df_reuters.loc[:, df_reuters.columns[-1]].value_counts().apply(lambda x: x < 200)]
print(f"There is {len(ids_to_remove)} documents with <= 200 samples in dataset.")
```

Remove documents which have less or equal samples than 200 in the entire dataset.

```python
df_reuters = df_reuters.groupby(30).filter(lambda x : len(x) > 200)
document_identifiers = df_reuters.iloc[:, -1].unique()
df_reuters.to_csv('data/reuters_reduced.csv', header=False, index=False)
print(f"Currently, Reuters dataset has samples for {len(document_identifiers)} unique documents. \n")
document_identifiers
```
